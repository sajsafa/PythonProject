## Aggregating Data with transform()


```python
orders_products_merge['dep_mean_order'] = orders_products_merge.groupby(['department_id'])['order_number'].transform(np.mean)
```

```python
orders_products_merge.groupby('department_id').agg({'order_number': ['mean']})
```

```python
df = orders_products_merge[:1000000]
```

In some cases, differences are more noticeable. in general, it can be concluded  that in "historical datasets", by limiting the sample space ( making a subset), the accuracy of statistical calculations decreases. So it should be better to make these calculations based on the entire data.

## Deriving Columns with loc()


```python
orders_products_merge.loc[orders_products_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
orders_products_merge.loc[(orders_products_merge['max_order'] <= 40) & (orders_products_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
orders_products_merge.loc[orders_products_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
orders_products_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
orders_products_merge[['user_id', 'loyalty_flag', 'order_number']].head()
```

```python
# 5.The marketing team at Instacart wants to know whether there’s a difference between the spending habits of the three types of customers you identified
```

```python
orders_products_merge.groupby('loyalty_flag').agg({'prices':['mean' , 'max', 'min']})
```

##### Based on the basic statistics above, the average amount paid by the Loyal Customer customer group is lower than the other groups, and as a result, there is no direct relationship between these two factors. There is no positive Correlation between loyalty and payment. Max and Min amounts in the three groups are the same.

```python
# 6.The team now wants to target different types of spenders in their marketing campaigns. This can be achieved by looking at the prices of the items people are buying
```

```python
orders_products_merge['user_mean_payment'] = orders_products_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
orders_products_merge[['user_id', 'loyalty_flag', 'user_mean_payment']].head()
```

```python
orders_products_merge.loc[orders_products_merge['user_mean_payment'] > 10, 'user_spender_flag'] = 'High spender'
orders_products_merge.loc[orders_products_merge['user_mean_payment'] <= 10, 'user_spender_flag'] = 'Low spender'
```

```python
orders_products_merge[['user_id', 'loyalty_flag', 'user_mean_payment','user_spender_flag']].iloc[70:80]
```

```python
#7.the Instacart team wants you to determine frequent versus non-frequent customers. Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column
```

```python
orders_products_merge['user_median_order'] = orders_products_merge.groupby('user_id')['days_since_prior_order'].transform(np.median)
```

```python
orders_products_merge[['user_id', 'days_since_prior_order', 'user_median_order']].iloc[:20]
```

```python
orders_products_merge.loc[orders_products_merge['user_median_order'] >20 ,'order_frequency_flag'] = 'Non-frequent customer'
orders_products_merge.loc[(orders_products_merge['user_median_order'] <= 20) & (orders_products_merge['user_median_order'] >10) ,'order_frequency_flag'] = 'Regular customer'
orders_products_merge.loc[orders_products_merge['user_median_order'] <= 10 ,'order_frequency_flag'] = 'Frequent customer'
```

```python
orders_products_merge[['user_id', 'user_median_order', 'order_frequency_flag']].iloc[95:100]
```

```python
orders_products_merge['order_frequency_flag'].value_counts(dropna = False)
```
